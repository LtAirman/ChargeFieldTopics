### Charge field Atoms

Charge generally enters a spinning proton's poles and 
is emitted closer to the proton's equator. A vertical 
spin axis aligned proton may be spinning either left or 
right (L or R). The electron can then orbit either the 
proton's top or bottom (T or B) pole. Some of the charge 
approaching the proton pole is partially blocked by the 
electron, resulting in a low charge field pressure zone 
'hole' near the proton's opposite pole. 

Bonds between protons occur when a proton occupies 
the low c.f. 'hole' of another ep pair proton. The 
surrounding c.f. keeps the protons together. Bonds may 
be in-line, such as in the up to six protons in a single 
stack sharing the same spin axis and charge channel; 
or orthogonal, where the equatorial emission plane of 
the proton (or stack) occupying the hole, is oriented 
sideways so as to directly feed the charge intake of 
the next proton pole (or stack end). 
                                   
All c.f. atoms are comprised of 1-19 stacks of 1-6 
electron proton (ep) pairs. Our working assumption is 
that Atoms within a molecule will follow the same 
bonding rules.

```python
"""Rotates and returns a tuple (x, y, z) of the 
three x, y, z positions and spin axis angle entered.  
ax, ay or az are angles in radians about the three
x y and z cubic spin axes through (0,0,0).
"""
def rotateAbtX(x, y, z, ax):
    rotatedX = x
    rotatedY = (y * cos(ax)) - (z * sin(ax))
    rotatedZ = (y * sin(ax)) + (z * cos(ax))
    x, y, z = rotatedX, rotatedY, rotatedZ    
    return(rotatedX, rotatedY, rotatedZ )

def rotateAbtY(x, y, z, ay):
    rotatedX = (z * sin(ay)) + (x * cos(ay))
    rotatedY = y
    rotatedZ = (z * cos(ay)) - (x * sin(ay))
    x, y, z = rotatedX, rotatedY, rotatedZ    
    return(rotatedX, rotatedY, rotatedZ )

def rotateAbtZ(x, y, z, az):
    rotatedX = (x * cos(az)) - (y * sin(az))
    rotatedY = (x * sin(az)) + (y * cos(az))
    rotatedZ = z
    x, y, z = rotatedX, rotatedY, rotatedZ    
    return(rotatedX, rotatedY, rotatedZ )
```

```python
""" 
Read and assemble atomic particle and hole locations 
from both Elements-Positions and EPNStacks for plotting  
"""

# The following two sets of lists, the first for iloc 
# and the second for loc indexing (not used), point to 
# the dfepn data frame columns containing proton electron 
# neutron and hole (x,y,z) coordinates for a given stack.

# dfepn particle coordinate columns for iloc indexing.
proXColList = [48, 54, 42, 60, 36, 66]
proYColList = [49, 55, 43, 61, 37, 67]
proZColList = [50, 56, 44, 62, 38, 68]
neuXColList = [80, 83, 86, 89, 92, 95]
neuYColList = [81, 84, 87, 90, 93, 96]
neuZColList = [82, 85, 88, 91, 94, 97]
eleXColList = [99, 102, 105, 108, 111, 114]
eleYColList = [100,103, 106 , 109, 112, 115]
eleZColList = [101, 104, 107, 110, 113, 116]

# Below, the 7 coordinate list allows up to 7 holes.
# There are only 1 or 2 holes in any stack.
# They can be re-sequenced later
holXColList = [51, 45, 57, 39, 63, 33, 69]
holYColList = [52, 46, 58, 40, 64, 34, 70]
holZColList = [53, 47, 59, 41, 65, 35, 71]

# dfepn particle coordinate columns for loc indexing.
#proXColList = ['p1x', 'p2x', 'p3x', 'p4x', 'p5x', 'p6x']
#proYColList = ['p1y', 'p2y', 'p3y', 'p4y', 'p5y', 'p6y']
#proZColList = ['p1z', 'p2z', 'p3z', 'p4z', 'p5z', 'p6z']
#neuXColList = ['n1x', 'n2x', 'n3x', 'n4x', 'n5x', 'n6x']
#neuYColList = ['n1y', 'n2y', 'n3y', 'n4y', 'n5y', 'n6y']
#neuZColList = ['n1z', 'n2z', 'n3z', 'n4z', 'n5z', 'n6z']
#eleXColList = ['e1x', 'e2x', 'e3x', 'e4x', 'e5x', 'e6x']
#eleYColList = ['e1y', 'e2y', 'e3y', 'e4y', 'e5y', 'e6y']
#eleZColList = ['e1z', 'e2z', 'e3z', 'e4z', 'e5z', 'e6z']
#holXColList = ['h1x', 'h2x', 'h3x', 'h4x', 'h5x', 'h6x', 'h7x']
#holXColList = ['h1y', 'h2y', 'h3y', 'h4y', 'h5y', 'h6y', 'h7y']
#holXColList = ['h1z', 'h2z', 'h3z', 'h4z', 'h5z', 'h6z', 'h7z']

# To contain all proton x, y, z coordinates.
allPSpinXs = []
allPSpinYs = []
allPSpinZs = []
# Proton Left spinners.
allPLSpinXs = []
allPLSpinYs = []
allPLSpinZs = []
# Proton Right spinners.
allPRSpinXs = []
allPRSpinYs = []
allPRSpinZs = []

# Proton oriented emissions.
allPOrien = [] # The value of SlotSpin: 'X', 'Y' or 'Z' values.
allNumPs = [] # Num of protons in the slot, to color-code emissions
    
# To contain all electron x, y, z coordinates.
allEXs = []
allEYs = []
allEZs = []
# Electrons near the Top Proton pole.
allETPosXs = []
allETPosYs = []
allETPosZs = []
# Electrons near the Bottom Proton pole.
allEBPosXs = []
allEBPosYs = []
allEBPosZs = []
ezValues = ['e1z', 'e2z', 'e3z', 'e4z', 'e5z', 'e6z']

#hzValues = ['h1z', 'h2z', 'h3z', 'h4z', 'h5z', 'h6z', 'h7z']
# To contain all hole x, y, z coordinates.
allHolXs = []
allHolYs = []
allHolZs = []

# These lists will contain all neutron coordinates. 
# All Neutrons. 'e1z', 'e1z', 'e1z', 
allNXs = []
allNYs = []
allNZs = []

# Neutrons need more thought
# Neutrons near the Top Proton pole.
#allNTPosXs = []
#allNTPosYs = []
#allNTPosZs = []
# Neutrons near the Bottom Proton pole.
#allNBPosXs = []
#allNBPosYs = []
#allNBPosZs = []

# Two functions that identify all protons by their coordinates
# (w.r.t. the slot center) and left or right spin (L or R) type.
def spinLeft(pXvalue, pYvalue, pZvalue, allPLSpinXs, allPLSpinYs, allPLSpinZs):
    allPLSpinXs.append(pXvalue) 
    allPLSpinYs.append(pYvalue)
    allPLSpinZs.append(pZvalue)
    return(allPLSpinXs, allPLSpinYs, allPLSpinZs) 

def spinRight(pXvalue, pYvalue, pZvalue, allPRSpinXs, allPRSpinYs, allPRSpinZs):
    allPRSpinXs.append(pXvalue) 
    allPRSpinYs.append(pYvalue)
    allPRSpinZs.append(pZvalue)
    return(allPRSpinXs, allPRSpinYs, allPRSpinZs)

# Two functions that identify all electrons by their proton 
# pole (T or B) position and coordinates (w.r.t the proton). 
def botElectron(pXvalue, pYvalue, pZvalue, allEBPosXs, allEBPosYs, allEBPosZs):                      
    allEBPosXs.append(eXvalue) 
    allEBPosYs.append(eYvalue) 
    allEBPosZs.append(eZvalue) 
    return(allEBPosXs, allEBPosYs, allEBPosZs)

def topElectron(pXvalue, pYvalue, pZvalue, allETPosXs, allETPosYs, allETPosZs):                      
    allETPosXs.append(eXvalue) 
    allETPosYs.append(eYvalue) 
    allETPosZs.append(eZvalue) 
    return(allETPosXs, allETPosYs, allETPosZs)

#InclHoles = False 
# For each of the 1-19 epn stacks for the selected atom
# contained in slotsetA_df. 
for j in range( 19 ): 
    
    # Only slots with  'CanBind' = True, will include holes
    InclHoles = False
    SingleHole = False
    TwoHoles = False
    if slotsetA_df.loc[j+1,('CanBind')] == True :
        InclHoles = True 
    
    # if the slot is occupied.    
    if slotsetA_df.loc[j+1,('Protons')] > 0 :
        numPs = slotsetA_df.loc[j+1,('Protons')] 
        
        #epnStack = int(slotsetA_df.loc[j+1,('StackNum')]) 
        epnStack = round(slotsetA_df.loc[j+1,('StackNum')]) 

        # For each epn in that slot   
        for k in range(numPs):
            
            allNumPs.append(numPs) 
            
            # Pointing to the number of N's in that set of numPs  
            # dfepn epns, N1,N2, N3, N4, N5, N6. 
            #intValue = slotsetA_df.iloc[j,intColumn]
            #intColumn = 74+k     
            intValue = dfepn.iloc[epnStack-1, 74+k]  # OK   
            #print('intValue = ',intValue)
            
            # The kth of numpy epn positions w.r.t. the slot.
            pXvalue = 0
            pYvalue = 0
            pZvalue = 0
            eXvalue = 0
            eYvalue = 0
            eZvalue = 0
            nXvalue = 0  
            nYvalue = 0
            nZvalue = 0
            n2Xvalue = 0  
            n2Yvalue = 0
            n2Zvalue = 0
            hXvalue = 0
            hYvalue = 0
            hZvalue = 0  
            h2Xvalue = 0
            h2Yvalue = 0
            h2Zvalue = 0          

            # All the epn particles and holes contained in EPNStacks 
            # are listed with a z-orientation. They need to be rotated
            # when the destination slot has an x or y orientation.
            z_to_y = -PI_PHI/2 # -90 deg rotation
            z_to_x = PI_PHI/2            

            # The stack holes are added to the stack during the k=numPs
            # loop, before the final epn particles are added below. 
            # Unlike electrons and neutrons which are located with respect  
            # to the proton poles they orbit, Holes have their own stack 
            # coordinates near the t and/or b stack ends.
            # Hloc values (‘h1’,‘h2’,‘h23’,‘h3’,‘h4’,‘h43’,‘h45’,‘h5’,‘h6’,‘h65’,‘h67’,‘h7’)
            # identify from 'Hloc' whether the stack has 1 or 2 of the 7 possible hole 
            # locations ('h1','h2','h3','h4','h5','h6','h7').
            if (InclHoles == True and k == numPs-1) :  # Add the holes when k = numPs

                if dfepn.loc[epnStack-1,'Hloc'] == 'h1' : 
                    SingleHole = True
                    # The single stack hole is at h1 (0f h1-h7) 
                    hXvalue = dfepn.loc[epnStack-1,'h1x']
                    hYvalue = dfepn.loc[epnStack-1,'h1y']
                    hZvalue = dfepn.loc[epnStack-1,'h1z']
 
                if dfepn.loc[epnStack-1,'Hloc'] == 'h2' : 
                    SingleHole = True
                    # The single stack hole is at h2
                    hXvalue = dfepn.loc[epnStack-1,'h2x']
                    hYvalue = dfepn.loc[epnStack-1,'h2y']
                    hZvalue = dfepn.loc[epnStack-1,'h2z']                       

                if dfepn.loc[epnStack-1,'Hloc'] == 'h3' :
                    SingleHole = True
                    # The single stack hole is at h3
                    hXvalue = dfepn.loc[epnStack-1,'h3x']
                    hYvalue = dfepn.loc[epnStack-1,'h3y']
                    hZvalue = dfepn.loc[epnStack-1,'h3z']                       

                if dfepn.loc[epnStack-1,'Hloc'] == 'h4' :
                    SingleHole = True
                    # The single stack hole is at h4
                    hXvalue = dfepn.loc[epnStack-1,'h4x']
                    hYvalue = dfepn.loc[epnStack-1,'h4y']
                    hZvalue = dfepn.loc[epnStack-1,'h4z']                       

                if dfepn.loc[epnStack-1,'Hloc'] == 'h5' :
                    SingleHole = True
                    # The single stack hole is at h5
                    hXvalue = dfepn.loc[epnStack-1,'h5x']
                    hYvalue = dfepn.loc[epnStack-1,'h5y']
                    hZvalue = dfepn.loc[epnStack-1,'h5z']                       

                if dfepn.loc[epnStack-1,'Hloc'] == 'h6' :
                    SingleHole = True
                    # Position the single stack hole at h6
                    hXvalue = dfepn.loc[epnStack-1,'h6x']
                    hYvalue = dfepn.loc[epnStack-1,'h6y']
                    hZvalue = dfepn.loc[epnStack-1,'h6z']                       

                if dfepn.loc[epnStack-1,'Hloc'] == 'h7' :
                    SingleHole = True
                    # The single stack hole is at h7
                    hXvalue = dfepn.loc[epnStack-1,'h7x']
                    hYvalue = dfepn.loc[epnStack-1,'h7y']
                    hZvalue = dfepn.loc[epnStack-1,'h7z']                       

                # 2 hole Hloc values ( h23’, ‘h43’, ‘h45’, ‘h65’, ‘h67’ )      
                if dfepn.loc[epnStack-1,'Hloc'] == 'h23' :
                    TwoHoles = True
                    # The 2 stack holes are at h2 and h3
                    hXvalue = dfepn.loc[epnStack-1,'h2x']
                    hYvalue = dfepn.loc[epnStack-1,'h2y']
                    hZvalue = dfepn.loc[epnStack-1,'h2z']                       
                    h2Xvalue = dfepn.loc[epnStack-1,'h3x']
                    h2Yvalue = dfepn.loc[epnStack-1,'h3y']
                    h2Zvalue = dfepn.loc[epnStack-1,'h3z']                       

                if dfepn.loc[epnStack-1,'Hloc'] == 'h43' :
                    TwoHoles = True
                   # The 2 stack holes are at h4 and h3
                    hXvalue = dfepn.loc[epnStack-1,'h4x']
                    hYvalue = dfepn.loc[epnStack-1,'h4y']
                    hZvalue = dfepn.loc[epnStack-1,'h4z']                       
                    h2Xvalue = dfepn.loc[epnStack-1,'h3x']
                    h2Yvalue = dfepn.loc[epnStack-1,'h3y']
                    h2Zvalue = dfepn.loc[epnStack-1,'h3z']                       
 
                if dfepn.loc[epnStack-1,'Hloc'] == 'h45' :
                    TwoHoles = True
                    # The 2 stack holes are at h4 and h5
                    hXvalue = dfepn.loc[epnStack-1,'h4x']
                    hYvalue = dfepn.loc[epnStack-1,'h4y']
                    hZvalue = dfepn.loc[epnStack-1,'h4z']                       
                    h2Xvalue = dfepn.loc[epnStack-1,'h5x']
                    h2Yvalue = dfepn.loc[epnStack-1,'h5y']
                    h2Zvalue = dfepn.loc[epnStack-1,'h5z']                       

                if dfepn.loc[epnStack-1,'Hloc'] == 'h65' :
                    TwoHoles = True
                    # The 2 stack holes are at h5 and h6
                    hXvalue = dfepn.loc[epnStack-1,'h6x']
                    hYvalue = dfepn.loc[epnStack-1,'h6y']
                    hZvalue = dfepn.loc[epnStack-1,'h6z']                       
                    h2Xvalue = dfepn.loc[epnStack-1,'h5x']
                    h2Yvalue = dfepn.loc[epnStack-1,'h5y']
                    h2Zvalue = dfepn.loc[epnStack-1,'h5z']                       

                if dfepn.loc[epnStack-1,'Hloc'] == 'h67' :
                    TwoHoles = True
                    # The 2 stack holes are at h6 and h7
                    hXvalue = dfepn.loc[epnStack-1,'h6x']
                    hYvalue = dfepn.loc[epnStack-1,'h6y']
                    hZvalue = dfepn.loc[epnStack-1,'h6z']                      
                    h2Xvalue = dfepn.loc[epnStack-1,'h7x']
                    h2Yvalue = dfepn.loc[epnStack-1,'h7y']
                    h2Zvalue = dfepn.loc[epnStack-1,'h7z']                              
                
                #'Rotate Holes to SlotOrien'
                if slotsetA_df.loc[j+1,('SlotOrien')] == "X" :
                    hXvalue, hYvalue, hZvalue = rotateAbtY(hXvalue, hYvalue, hZvalue, z_to_x)
                    if TwoHoles == True :
                        h2Xvalue, h2Yvalue, h2Zvalue = rotateAbtY(h2Xvalue, h2Yvalue, h2Zvalue, z_to_x)       

                if slotsetA_df.loc[j+1,('SlotOrien')] == "Y" :
                    hXvalue, hYvalue, hZvalue = rotateAbtX(hXvalue, hYvalue, hZvalue,z_to_y)
                    if TwoHoles == True :
                        h2Xvalue, h2Yvalue, h2Zvalue = rotateAbtX(h2Xvalue, h2Yvalue, h2Zvalue, z_to_y)      

                # Copied from the proton section below, 
                #'PX','PY','PZ' identify each of the 1-19 SL slot center
                # coordinates with respect to the atom's center at (0,0,0). 
                # Slot1's center coincides with the atom's center. 
                hXvalue += slotsetA_df.loc[j+1,('PX')]
                hYvalue += slotsetA_df.loc[j+1,('PY')]
                hZvalue += slotsetA_df.loc[j+1,('PZ')]       
                hXvalue = round(hXvalue,4) 
                hYvalue = round(hYvalue,4)
                hZvalue = round(hZvalue,4)

                if TwoHoles == True :
                    h2Xvalue += slotsetA_df.loc[j+1,('PX')]
                    h2Yvalue += slotsetA_df.loc[j+1,('PY')]
                    h2Zvalue += slotsetA_df.loc[j+1,('PZ')]          
                    h2Xvalue = round(h2Xvalue,4)
                    h2Yvalue = round(h2Yvalue,4)
                    h2Zvalue = round(h2Zvalue,4)                

                # To contain all hole x, y, z coordinates.
                allHolXs.append(hXvalue)
                allHolYs.append(hYvalue)
                allHolZs.append(hZvalue)
                if TwoHoles == True :
                    allHolXs.append(h2Xvalue)
                    allHolYs.append(h2Yvalue)
                    allHolZs.append(h2Zvalue)           
                
                # The holes have been added                      
                
            # Protons. 
            pXvalue = dfepn.iloc[epnStack-1,(proXColList[k])]
            pYvalue = dfepn.iloc[epnStack-1,(proYColList[k])]
            pZvalue = dfepn.iloc[epnStack-1,(proZColList[k])]
            #print('pXvalue = ', pXvalue,' pYvalue = ', pYvalue,' pZvalue = ', pZvalue)

            #'Rotate Protons to SlotOrien'
            if slotsetA_df.loc[j+1,('SlotOrien')] == "X" :
                #print('Rotate Protons to X')   
                pXvalue, pYvalue, pZvalue = rotateAbtY(pXvalue, pYvalue, pZvalue,z_to_x)   
                
            if slotsetA_df.loc[j+1,('SlotOrien')] == "Y" :
                #print('Rotate Protons to Y')   
                pXvalue, pYvalue, pZvalue = rotateAbtX(pXvalue, pYvalue, pZvalue,z_to_y)   
            
            #'PX','PY','PZ' identify each of the 1-19 SL slot center 
            # coordinates with respect to the atom's center at (0,0,0). 
            # Slot1's center coincides with the atom's center. 
            pXvalue += slotsetA_df.loc[j+1,('PX')]
            pYvalue += slotsetA_df.loc[j+1,('PY')]
            pZvalue += slotsetA_df.loc[j+1,('PZ')]            
            
            pXvalue = round(pXvalue,4)
            pYvalue = round(pYvalue,4)
            pZvalue = round(pZvalue,4)
            
            # All dfepn stacks have been left z-oriented, or rotated to 
            # agree with the slotsetA_df 'SlotOrien' 'X' or 'Y' directions. 
            #allPOrien.append(slotsetA_df.loc[j+1,('SlotOrien')])

            # The protons were entered directly from the data frame. 
            # Electrons and neutrons will be given random pos about Z first.
            eXvalue = dfepn.iloc[epnStack-1,(eleXColList[k])] 
            eYvalue = dfepn.iloc[epnStack-1,(eleYColList[k])] 
            eZvalue = dfepn.iloc[epnStack-1,(eleZColList[k])]
            
            #'Rotate Electrons to SlotOrien'            
            if slotsetA_df.loc[j+1,('SlotOrien')] == "X" :
                #print('Rotate Electrons to X')   
                eXvalue, eYvalue, eZvalue = rotateAbtY(eXvalue, eYvalue, eZvalue, z_to_x)              
 
            if slotsetA_df.loc[j+1,('SlotOrien')] == "Y" :
                #print('Rotate Electrons to Y')   
                eXvalue, eYvalue, eZvalue = rotateAbtX(eXvalue, eYvalue, eZvalue, z_to_y)             
            
            #Before randomizing assume there are neutron pairs so that 
            #they will remain 180deg apart.
            # Z's must be handled differently, depending on whether 
            # there are 0,1 or2 neutrons in a single proton set.
            # Z axis aligned.
            
            nXvalue = dfepn.iloc[epnStack-1,(neuXColList[k])]  
            nYvalue = dfepn.iloc[epnStack-1,(neuYColList[k])]
            nZvalue = dfepn.iloc[epnStack-1,(neuZColList[k])]
            
            #'Rotate Neutrons to SlotOrien'            
            if slotsetA_df.loc[j+1,('SlotOrien')] == "X" :
                #print('Rotate Neutrons to X')   
                nXvalue, nYvalue, nZvalue = rotateAbtY(nXvalue, nYvalue, nZvalue, z_to_x)                          
   
            if slotsetA_df.loc[j+1,('SlotOrien')] == "Y" :
                #print('Rotate Neutrons to Y')   
                nXvalue, nYvalue, nZvalue = rotateAbtX(nXvalue, nYvalue, nZvalue, z_to_y)   
            
            #The protons have been appended. Give 
            #the e and 1 or 2 n's arbitrary orbital positions 
            #about the proton's spin axis. Before randomizing 
            #the neutron's position, add a second neutron on 
            #the opposite side z-axis proton spin axis. The 2nd
            #neuton will not be appended to the particle 
            #list if it is not needed.
            
            # For randomizing neutrons and electrons within their orbits. 
            ax = random()*PI_PHI    
            ay = random()*PI_PHI
            az = random()*PI_PHI    
            neu_anglex = random()*PI_PHI  
            neu_angley = random()*PI_PHI          
            neu_anglez = random()*PI_PHI              
            
            # Append electron coords to allEXs  
            eXvalue = round(eXvalue + pXvalue,4)
            eYvalue = round(eYvalue + pYvalue,4) 
            eZvalue = round(eZvalue + pZvalue,4) 
            
            allEXs.append(eXvalue) 
            allEYs.append(eYvalue) 
            allEZs.append(eZvalue)
            
            # 'SlotSpin' (and 'Electrons') values include:
            # Four constant e-pos and p-spin. TRTR,BRBR,BLBL,TLTL
            # Eight alpha types A, V, E, W, M, C, P, Q, L, R, 
            # TB (electron position) alpha types:
            # A. TLbtBR. L p-spin, T e-pos top and R spinning bottom proton(s).
            # V. TRbtBR. R spinning top and L spinning bottom proton(s).
            # E. TRbtBR. R spinning proton(s).
            # W. TLbtBL. L spinning proton(s).
            # BT (electron position) alpha types: 
            # P. tBLTRb. L spinning top and R spinning bottom proton(s).
            # C. tBRTRb. R spinning proton(s).
            # M. tBLTLb. L spinning proton(s).
            # Q. tBRTLb. R spinning top and L spinning bottom proton(s). 
            # N is an empty slot.    
            
            # For proton equatorial emisssions
            # All dfepn stacks have been left z-oriented, or rotated to 
            # agree with the slotsetA_df 'SlotOrien' 'X' or 'Y' directions. 

            allPSpinXs.append(pXvalue) 
            allPSpinYs.append(pYvalue) 
            allPSpinZs.append(pZvalue)
            allPOrien.append(slotsetA_df.loc[j+1,('SlotOrien')])
            
            if dfepn.loc[epnStack-1,('SlotSpin')] == 'A':
                # A. TLbtBR.
                if dfepn.loc[epnStack-1,ezValues[k]] > 0 : 
                    spinLeft(pXvalue, pYvalue, pZvalue, allPLSpinXs, allPLSpinYs, allPLSpinZs)                    
                    topElectron(pXvalue, pYvalue, pZvalue, allETPosXs, allETPosYs, allETPosZs)  
                # A type alpha bottom.
                elif dfepn.loc[epnStack-1,ezValues[k]] < 0 :
                    spinRight(pXvalue, pYvalue, pZvalue, allPRSpinXs, allPRSpinYs, allPRSpinZs)                    
                    botElectron(pXvalue, pYvalue, pZvalue, allEBPosXs, allEBPosYs, allEBPosZs)                 

            if dfepn.loc[epnStack-1,('SlotSpin')] == 'V':
                # V. TRbtBR. 
                if dfepn.loc[epnStack-1,ezValues[k]] > 0 : 
                    spinRight(pXvalue, pYvalue, pZvalue, allPRSpinXs, allPRSpinYs, allPRSpinZs)                                       
                    topElectron(pXvalue, pYvalue, pZvalue, allETPosXs, allETPosYs, allETPosZs)  
                elif dfepn.loc[epnStack-1,ezValues[k]] < 0 :
                    spinLeft(pXvalue, pYvalue, pZvalue, allPLSpinXs, allPLSpinYs, allPLSpinZs)
                    botElectron(pXvalue, pYvalue, pZvalue, allEBPosXs, allEBPosYs, allEBPosZs)          

            if dfepn.loc[epnStack-1,('SlotSpin')] == 'E':
                # E. TRbtBR. 
                spinRight(pXvalue, pYvalue, pZvalue, allPRSpinXs, allPRSpinYs, allPRSpinZs)                    
                if dfepn.loc[epnStack-1,ezValues[k]] > 0 :                      
                    topElectron(pXvalue, pYvalue, pZvalue, allETPosXs, allETPosYs, allETPosZs)  
                elif dfepn.loc[epnStack-1,ezValues[k]] < 0 : 
                    botElectron(pXvalue, pYvalue, pZvalue, allEBPosXs, allEBPosYs, allEBPosZs)

            if dfepn.loc[epnStack-1,('SlotSpin')] == 'W': 
                # W. TLbtBL. 
                spinLeft(pXvalue, pYvalue, pZvalue, allPLSpinXs, allPLSpinYs, allPLSpinZs)
                if dfepn.loc[epnStack-1,ezValues[k]] < 0 :
                    topElectron(pXvalue, pYvalue, pZvalue, allETPosXs, allETPosYs, allETPosZs)  
                elif dfepn.loc[epnStack-1,ezValues[k]] > 0 :
                    botElectron(pXvalue, pYvalue, pZvalue, allEBPosXs, allEBPosYs, allEBPosZs)
                    
            if dfepn.loc[epnStack-1,('SlotSpin')] == 'L': 
                # TLTLb or tBLBL
                spinLeft(pXvalue, pYvalue, pZvalue, allPLSpinXs, allPLSpinYs, allPLSpinZs)
                #allPOrien.append(dfepn.loc[epnStack-1,('SlotOrien')])
                if dfepn.loc[epnStack-1,('Electrons')] == "T":
                    topElectron(pXvalue, pYvalue, pZvalue, allETPosXs, allETPosYs, allETPosZs)  
                elif dfepn.loc[epnStack-1,('Electrons')] == "B":
                    botElectron(pXvalue, pYvalue, pZvalue, allEBPosXs, allEBPosYs, allEBPosZs)           
          
            if dfepn.loc[epnStack-1,('SlotSpin')] == 'R':
                # TRTRb or tBRBR
                spinRight(pXvalue, pYvalue, pZvalue, allPRSpinXs, allPRSpinYs, allPRSpinZs)
                if dfepn.loc[epnStack-1,('Electrons')] == "T":
                    topElectron(pXvalue, pYvalue, pZvalue, allETPosXs, allETPosYs, allETPosZs)  
                elif dfepn.loc[epnStack-1,('Electrons')] == "B":
                    botElectron(pXvalue, pYvalue, pZvalue, allEBPosXs, allEBPosYs, allEBPosZs)        
            
            if dfepn.loc[epnStack-1,('SlotSpin')] == 'C':
                # C. tBRTRb. R spinning proton(s). 
                spinRight(pXvalue, pYvalue, pZvalue, allPRSpinXs, allPRSpinYs, allPRSpinZs)                       
                if dfepn.loc[epnStack-1,ezValues[k]] < 0 :    
                    botElectron(pXvalue, pYvalue, pZvalue, allEBPosXs, allEBPosYs, allEBPosZs)
                elif dfepn.loc[epnStack-1,ezValues[k]] > 0 :                     
                    topElectron(pXvalue, pYvalue, pZvalue, allETPosXs, allETPosYs, allETPosZs)  

            if dfepn.loc[epnStack-1,('SlotSpin')] == 'M':
                # M. tBLTLb. L spinning proton(s).
                spinLeft(pXvalue, pYvalue, pZvalue, allPLSpinXs, allPLSpinYs, allPLSpinZs)
                if dfepn.loc[epnStack-1,ezValues[k]] < 0 :    
                    botElectron(pXvalue, pYvalue, pZvalue, allEBPosXs, allEBPosYs, allEBPosZs)
                elif dfepn.loc[epnStack-1,ezValues[k]] > 0 : 
                    topElectron(pXvalue, pYvalue, pZvalue, allETPosXs, allETPosYs, allETPosZs)   
            
            if dfepn.loc[epnStack-1,('SlotSpin')] == 'P':
                # P tBLTRb.    
                if dfepn.loc[epnStack-1,ezValues[k]] < 0 :                       
                    spinLeft(pXvalue, pYvalue, pZvalue, allPLSpinXs, allPLSpinYs, allPLSpinZs)
                    botElectron(pXvalue, pYvalue, pZvalue, allEBPosXs, allEBPosYs, allEBPosZs)
                elif dfepn.loc[epnStack-1,ezValues[k]] > 0 :   
                    spinRight(pXvalue, pYvalue, pZvalue, allPRSpinXs, allPRSpinYs, allPRSpinZs)                    
                    topElectron(pXvalue, pYvalue, pZvalue, allETPosXs, allETPosYs, allETPosZs)           
                
            if dfepn.loc[epnStack-1,('SlotSpin')] == 'Q':  
                #Q. tBRTLb.           
                if dfepn.loc[epnStack-1,ezValues[k]] < 0 : 
                    spinRight(pXvalue, pYvalue, pZvalue, allPRSpinXs, allPRSpinYs, allPRSpinZs)                    
                    botElectron(pXvalue, pYvalue, pZvalue, allEBPosXs, allEBPosYs, allEBPosZs)
                elif dfepn.loc[epnStack-1,ezValues[k]] > 0 :               
                    spinLeft(pXvalue, pYvalue, pZvalue, allPLSpinXs, allPLSpinYs, allPLSpinZs)
                    topElectron(pXvalue, pYvalue, pZvalue, allETPosXs, allETPosYs, allETPosZs) 
            
            if slotsetA_df.loc[j+1,('SlotOrien')] == "X" :
                #print('Rotate X')   
                n2Xvalue = nXvalue
                n2Yvalue = nYvalue
                n2Zvalue = -nZvalue
                eXvalue,eYvalue,eZvalue = rotateAbtX(eXvalue,eYvalue,eZvalue,ax)
                nXvalue,nYvalue,nZvalue = rotateAbtX(nXvalue,nYvalue,nZvalue,neu_anglex) 
                n2Xvalue,n2Yvalue,n2Zvalue = rotateAbtX(n2Xvalue,n2Yvalue,n2Zvalue,neu_anglex) 
                     
            if slotsetA_df.loc[j+1,('SlotOrien')] == "Y" :
                n2Xvalue = -nXvalue
                n2Yvalue = nYvalue
                n2Zvalue = nZvalue            
                eXvalue,eYvalue,eZvalue = rotateAbtY(eXvalue,eYvalue,eZvalue,ay)            
                nXvalue,nYvalue,nZvalue = rotateAbtY(nXvalue,nYvalue,nZvalue,neu_angley) 
                n2Xvalue,n2Yvalue,n2Zvalue = rotateAbtY(n2Xvalue,n2Yvalue,n2Zvalue,neu_angley) 

            if slotsetA_df.loc[j+1,('SlotOrien')] == "Z" :
                n2Xvalue = -nXvalue 
                n2Yvalue = nYvalue
                n2Zvalue = nZvalue
                eXvalue,eYvalue,eZvalue = rotateAbtZ(eXvalue,eYvalue,eZvalue,az)          
                nXvalue,nYvalue,nZvalue = rotateAbtZ(nXvalue,nYvalue,nZvalue,neu_anglez) 
                n2Xvalue,n2Yvalue,n2Zvalue = rotateAbtZ(n2Xvalue,n2Yvalue,n2Zvalue,neu_anglez) 
                     
            nXvalue = round(nXvalue + pXvalue,4)
            nYvalue = round(nYvalue + pYvalue,4)
            nZvalue = round(nZvalue + pZvalue,4)
            n2Xvalue = round(n2Xvalue + pXvalue,4)
            n2Yvalue = round(n2Yvalue + pYvalue,4)
            n2Zvalue = round(n2Zvalue + pZvalue,4)
            
            # Add as many neutrons as the epn slot contains, 0,1 or 2.
            # 2nd will be opposite the first
            if intValue == 1 :
                allNXs.append(nXvalue) 
                allNYs.append(nYvalue) 
                allNZs.append(nZvalue)    
                #print()
            elif intValue == 2 :                
                allNXs.append(nXvalue) 
                allNYs.append(nYvalue) 
                allNZs.append(nZvalue)   
                allNXs.append(n2Xvalue) 
                allNYs.append(n2Yvalue) 
                allNZs.append(n2Zvalue)

```
